```python
import pandas as pd
import dowhy

# Avoiding unnecessary log messges and warnings
import logging
logging.getLogger("dowhy").setLevel(logging.WARNING)
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
```

```python
DF_SAVEFILENAME = '/home/vaisakh/developer/SysML/Project/PruningTest/LeNet5-experiment.csv'
data = pd.read_csv(DF_SAVEFILENAME, sep=",", encoding='utf-8')
```

```python
display(data['acc'])
data['High_Accuracy_Loss'] = data['acc'].apply(lambda x: True if x < 96.0 else False)

display(data)
```

```python
data['High_Pruning'] = data['prune_ratio'].apply(lambda x: True if x > 0.3 else False)
display(data)

data['High_Memory_Consumption'] = data['vm_percent'].apply(lambda x: True if x > 80.0 else False)
display(data)
```

```python
training= data[['High_Pruning','High_Accuracy_Loss','High_Memory_Consumption','prune_ratio','acc','memory_consumption','vm_percent','swap_percent','flops']].copy()

causal_graph = """
digraph {
High_Pruning;
High_Accuracy_Loss;
High_Memory_Consumption;
prune_ratio;
acc;
flops;
memory_consumption;
vm_percent;
swap_percent;
U[label="Unobserved Confounders"];
vm_percent->memory_consumption;swap_percent->memory_consumption;
memory_consumption->High_Memory_Consumption;vm_percent->High_Memory_Consumption;swap_percent->High_Memory_Consumption;
prune_ratio->High_Pruning;prune_ratio->memory_consumption;prune_ratio->flops;pruning_ration->acc;
acc->High_Accuracy_Loss;prune_ratio->High_Accuracy_Loss;High_Pruning->High_Accuracy_Loss;U->High_Accuracy_Loss;
}
"""


# Customer_Age -> Education_Level; Customer_Age -> Income_Category;
# Education_Level -> Income_Category; Income_Category->High_limit;
# U->Income_Category;U->High_limit;U->Churn;
# High_limit->Churn; Income_Category -> Churn;

from dowhy import CausalModel
from IPython.display import Image,display

model=CausalModel(
        data = training,
        graph=causal_graph.replace("\n", " "),
        treatment='High_Pruning',
        outcome='High_Accuracy_Loss')
model.view_model()
display(Image(filename="causal_model.png"))
```

```python
#Identify the causal effect
estimands = model.identify_effect()
print(estimands)
```

```python
#Causal Effect Estimation
estimate = model.estimate_effect(estimands,method_name = "backdoor.propensity_score_weighting")
print(estimate)
```

```python
refutel = model.refute_estimate(estimands,estimate, "random_common_cause")
print(refutel)
```

```python
refutel = model.refute_estimate(estimands,estimate, "data_subset_refuter")
print(refutel)
```

```python
refutel = model.refute_estimate(estimands,estimate, "placebo_treatment_refuter")
print(refutel)
```
